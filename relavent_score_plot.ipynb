```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "relevance_scores.xlsx" 
df = pd.read_excel(file_path)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df["Test Case"], df["Relevance Score (1536 Dimension)"], marker='o', label='1536 Dimension')
plt.plot(df["Test Case"], df["Relevance Score (768 Dimension)"], marker='s', label='768 Dimension')
plt.title('Relevance Score: 1536 vs 768 Embedding Dimensions')
plt.xlabel('Test Case')
plt.ylabel('Relevance Score')
plt.ylim(0.6, 1.0)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
